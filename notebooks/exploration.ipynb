We decided to drop the 'company' column since more than 80% of its values are missing.

Additionally, we filled the 'agent' column with a new ID, -1, to represent "None" in our context.

For the 'children' column, we replaced missing values with the mean of the column.

Finally, we removed rows where the 'country' value was null.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```

# Understading booking cancelations

```python
# Cancellation rate 1 means canceled
df['is_canceled'].value_counts(normalize=True) * 100
```

```python
# Cancellations by hotel
df.groupby('hotel')['is_canceled'].mean()
```

```python
# Trends
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df.groupby(df['reservation_status_date'].dt.to_period('M'))['is_canceled'].mean().plot(kind='line', figsize=(10,5), title="Cancellation Trends Over Time")
```

```python
df['customer_type'].value_counts()
```

```python
df.groupby('market_segment')['is_canceled'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Feature engineering
# Not important
# df['total_guests'] = df['adults'] + df['children'] + df['babies']
# df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
# df['price_per_person'] = df['adr'] / (df['total_guests'] + 1)  # Avoid division by zero

# Categorical to numerical

df['hotel'] = df['hotel'].astype('category').cat.codes
df['customer_type'] = df['customer_type'].astype('category').cat.codes
df['market_segment'] = df['market_segment'].astype('category').cat.codes

plt.figure(figsize=(15,8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
